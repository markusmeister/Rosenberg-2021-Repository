# Introduction
This notebook serves to reproduce the figures presented in our preprint on maze behavior:

Matthew Rosenberg, Tony Zhang, Pietro Perona, Markus Meister (2021) Mice in a labyrinth: Rapid learning, sudden insight, and efficient exploration


## Organization of the repo
All the specialty code used, other than common Python packages, is contained in the `/code` directory of this repo.  

An "experiment" is observation of one mouse in the maze for one night, typically a period of 7 h. There are 20 experiments, each from one mouse during its first encounter with the labyrinth. The mice are named 'B1', 'B2' etc. 

The raw video from an experiment was tracked using methods described in the preprint. From that emerges a "trajectory file" that contains all the information about the mouse's movements we need for analysis. These files live in the `/outdata` directory and are called 'B1-tf', 'B2-tf', etc. 

From the trajectory file we compute various intermediate results. These also go in the `/outdata` directory.

Ultimately we produce figures in PDF format. These go into the `/figs` directory. Most often the function producing a figure will start directly from the trajectory files. In a few cases it relies on an intermediate result.

These PDF figure panels were composed into figures using Keynote. No post-processing steps affected the content of the data.

## Data structures
There are only 2 structures of interest: a maze (`Maze`), and a trajectory (`Traj`). For their definitions look at the functions `NewMaze()` and `NewTraj()`. Each trajectory file is simply the pickled version of a trajectory structure. 


## Correspondence of code to figures
For every figure panel in the preprint (as of 1/13/2021) there is a level-3 heading in one of the four notebooks (A-D).

## Data scrubbing
All the data analysis is based on the trajectory of the nose keypoint through the maze. The raw data from video tracking were curated as follows:
- We manually corrected some tracking irregularities. A common failure occurred when the animal followed a wall with its nose, and the tracker occasionally placed the keypoint on the wrong side of a wall.
- We digitized the continuous trajectory by binning space into square cells the width of a corridor. When the keypoint crosses a boundary between cells it occasionally flickers back and forth rapidly. We deleted all such flickers that consisted of a single video frame (1/30 s).

The results are stored in a `Traj` structure and saved in the "trajectory files", named `B1-tf`, `B2-tf` etc.

## Draw nose trajectories

```python
# Sequentially draw every bout on top of the maze outline
def InspectNose(tr,ma):
    '''
    Sequentially plots every bout. Nose trajectory only.
    Enter:
    <return> to go to next bout
    'q' to quit
    'b' to mark this bout as bad
    '<int>' to move to bout number <int>
    Returns the list of bad bouts
    '''

    from IPython.display import clear_output

    def PlotNose(bo,tr,ma):
        '''
        Plot bout bo of trajectory tr in maze ma
        Both continuous and discrete positions
        Use an approximate scaling that seems to align them
        '''
        ax=PlotMazeWall(ma)
        plot(-0.5+15*tr.ke[bo][:,0],-0.5+15*tr.ke[bo][:,1],
                  fmts=['r-'],linewidth=1,figsize=(4,4),equal=True,yflip=True,
                  xlim=[-1,15],ylim=[-1,15],xhide=True,yhide=True,axes=ax);

    bad=[]
    i=0
    while i<len(tr.ce):
        print(i,flush=True)
        PlotNose(i,tr,ma)
        plt.show()
        a=input("enter: ")
        if a=='b':
            bad+=[i,]
        elif a=='q':
            break
        else:
            try:
                j=int(a)
            except:
                j=-1
            if j in range(len(tr.ce)):
                i=j-1    
        i=i+1
        clear_output(wait=True)
    return bad
    
```

## Speed
Start from the cell trajectory. Whenever the cell changes, record the time-in-maze. Then can histogram that event train at whatever resolution you like.

### Speed vs time for each animal 

```python
# compute and display speed for all animals
Names=RewNames+UnrewNamesSub
k=len(RewNames)
bw=100 # bin width in s
tm=15000 # max time to be analyzed
spa=np.full((len(Names),int(tm/bw)),-1,dtype=float) # array to hold all the speeds
for i,nickname in enumerate(Names):
    tf=LoadTraj(nickname+'-tf')
    # construct a point process of all the steps between adjacent cells
    ct=[] # step times, measured in time-in-maze
    for j,b in enumerate(tf.ce):
        d=b[:] # cell for each frame
        d[1:]-=d[:-1] # change in cell
        d[0]=1 # first cell called changed
        t0=TimeInMaze(tf.fr[j,0],tf) # time in maze for start of the bout
        ct+=(np.where(d>0)[0]/30+t0).tolist() # time in maze for the steps
    ct=np.array(ct)
    # histogram that point process to compute speed = d(steps)/d(time)
    nb=int(np.min((np.max(ct),tm))/bw) # num bins
    bins=np.linspace(0,nb*bw,nb+1,endpoint=True)
    sp,ed=np.histogram(ct,bins)
    sp=sp/bw
    spa[i,:len(sp)]=sp # add bins that are available
    print(nickname)
    hist(ed[:-1],ed,weights=sp,xlabel='Time in maze (s)',ylabel='Speed (cells/s)',
        figsize=(12,2),xlim=[0,tm]); 
    plt.show()
```

Speed in the maze is remarkably constant over time for all the animals. There are a few dips, probably for grooming and short naps. But no long term trends.

Also all speeds are 1-2 cells/s.

## Transitions between cage and maze


Look at the time intervals between a return to the cage and the next entry into the maze. Ignore the first entry because it requires discovery of the maze.

- is there any periodicity to entries into the maze
- does the probability increase with time spent in the cage, as if the drive to explore increases with time?

Think of return to maze as a renewal process with a rate per unit time $r(t)$ that depends on time spent in cage. Then the probability of entering at time $t$ is

$$p\left( t \right) = {e^{ - \int\limits_0^t {r\left( {t'} \right)dt'} }}r\left( t \right) =  - \frac{d}{{dt}}u\left( t \right)$$

where

$$u\left( t \right) = {e^{ - \int\limits_0^t {r\left( {t'} \right)dt'} }}$$

So

$$\int\limits_0^t {r\left( {t'} \right)dt'}  =  - \ln \left( {1 - \int\limits_0^t {p\left( {t'} \right)dt'} } \right)$$

This relates the cumulative of the rate function $r(t)$ to the cumulative of the inter-event distribution $p(t)$.

### Figure 1–figure supplement 3

#### Unrewarded animals entry into maze

```python
# Instantaneous rate of entry into maze, avg over all rewarded animals
Names=UnrewNamesSub
tc=[]
for i,nickname in enumerate(Names):
    tf=LoadTraj(nickname+'-tf')
    x=np.array([[b[0]/30,b[1]/30] for b in tf.fr]) # start and end of each bout in s
    tc+=[np.array(x[1:,0]-x[:-1,1])] # time in cage = start of bout minus end of previous bout
t=np.concatenate(tc) # time intervals in cage for all animals
t=np.sort(t) # sort them
p=(np.arange(len(t)+1))/len(t) # cumulative prob starting at zero
c=-np.log(1-p[:-1]) # log transform omit last element of count to avoid log(0), len(t) elements
d=c[1:]-c[:-1] # how much each entry contributes to cum rate, len(t)-1 elements
t=t[:-1] # omit last element of time
b=2 # binwidth in s
n=100 # number of bins
hist(t,weights=d/b,bins=np.arange(n+1)*b,ylabel='Instantaneous rate (1/s)',xlabel='Time in cage (s)',
    figsize=(7,3),xlim=[0,b*n],color=['g'],title='Unrewarded animals: Rate of entry into maze',loc='upper right',
    ylim=[0,0.046]);
plt.savefig('figs/Rate_Entry_Unrew_2.pdf');
```

The rate of entry to the maze is highest just after the animal returns to the cage.
